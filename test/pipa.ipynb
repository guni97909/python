```python
import pandas as pd 
Allstat=pd.read_csv("./PlayerStatisticR.csv",index_col=0,encoding='cp949')
Allstat = Allstat.reset_index(drop=True)
Allstat.head()   # 2022 프리미엄리그 선수들 스텟
```

```python
# 메흐디 타레미 아래를 보고싶음  5~9 선택적으로 자료를 불러올수 있음.
Allstat[5:10]
```

```python
# One으로 되어있는 데이터를 1로 바꾸자
Allstat=Allstat.replace('One','1')
# -표시를 0으로 교체
Allstat=Allstat.replace('-','0')
# NaN 은 numpy라서 새로 import 해야한다
import numpy as np
Allstat=Allstat.replace(np.nan,'0')
Allstat
```

```python
# apps 는 출전경기수인데 괄호표시는()  교체출전이기 때문에 이것을 고민 많이했는데.. 둘다 출전으로 합치고 minutes 출전시간을 나누면 되지 않을까?? 
# 일단 출전경기수  주전+교체 상관없이 출전횟수로 변경 
for i in range(len(Allstat)):
    if '(' in Allstat['apps'][i]:
        if '(' in Allstat['apps'][i][0:2]:
            if len(Allstat['apps'][i][2:])==2:
                #print(int(Allstat['apps'][i][0:1])+int(Allstat['apps'][i][2:3]))
                Allstat['apps'][i]=int(Allstat['apps'][i][0:1])+int(Allstat['apps'][i][2:3])
            else:
                #print(int(Allstat['apps'][i][0:1])+int(Allstat['apps'][i][2:4]))
                Allstat['apps'][i]=int(Allstat['apps'][i][0:1])+int(Allstat['apps'][i][2:4])
        else:
            #print(int(Allstat['apps'][i][0:2])+int(Allstat['apps'][i][3:4]))
            Allstat['apps'][i]=int(Allstat['apps'][i][0:2])+int(Allstat['apps'][i][3:4])
```

```python
from sklearn.linear_model import LinearRegression  
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

Allstat=Allstat.drop(['player'],axis=1) # 일단 분석을 해야하기 때문에  'player'   name은 제거
Allstat
```

```python
# X,Y 분활
Y=Allstat['evaluation']
X=Allstat.drop(['evaluation'],axis=1, inplace=False)
```

```python
# 훈련용 데이터와 평가용 데이터 분할
X_train,X_test,Y_train,y_test=train_test_split(X,Y, test_size=0.3, random_state=156)
```

```python
# 선형회귀 분석 : 모델생성
lr=LinearRegression()
```

```python
# 선형회귀 분석: 모델 훈련
lr.fit(X_train,Y_train)
```
